## df_lost_property - Properties That Have Lost Assisted Housing Status: Exploration

This dataset tracks **rental properties in Alachua County that have lost their assisted housing status** â€” meaning they no longer receive government or quasi-government support. Each row represents a single property that was previously part of the Assisted Housing Inventory but has since exited, whether through expired contracts, mortgage prepayment, or the end of use restrictions.

#### Column Descriptions

| Column | Description |
|--------|-------------|
| `LPI ID` | Lost Property Inventory identifier |
| `Shim ID` | Shimberg Center internal identifier |
| `FHFC HPP key` | Florida Housing Finance Corporation Housing Property Portfolio key |
| `Name` | Development name as reported by funding sources |
| `Address` | Street address |
| `City` | City |
| `Zip` | ZIP code |
| `County` | County (all Alachua in this subset) |
| `Total Units` | Total number of units in the development, including property management units |
| `Estimated Assisted Units` | Estimated number of units that had rent and/or income restrictions. For early LPI developments, this is an estimate: if the property appeared in the AHI previously, it reflects the last reported count; otherwise, it is the minimum required by the funding source (e.g., 20% of total units for Bonds/LIHTC/SAIL, 35% for FDIC). |
| `HUD/RD Rental Assistance Units` | Units that received monthly rental assistance from HUD or USDA Rural Development |
| `Housing Program(s)` | Program(s) that funded the property, administered by HUD, USDA RD, Florida Housing, or LHFA. Only programs with income and/or rent restrictions are reported; programs without restrictions are included only if combined with restricted programs. |
| `FHFC Funded` | Whether the property was funded by Florida Housing Finance Corporation |
| `RD Funded` | Whether the property was funded by USDA Rural Development |
| `LHFA Funded` | Whether the property was funded by a Local Housing Finance Authority |
| `HUD Funded` | Whether the property was funded by HUD |
| `Approximate Year of Loss` | Approximate year the property lost its assisted status. The data dictionary refers to this column as "Year of Loss" and defines it as the last year the property appeared in the AHI, or the latest of: date of last report to FHFC, year of opt-out from HUD rental assistance, or year of mortgage prepayment/assignment. |
| `Tax Credit or Bond Property No Longer Under Use Restrictions` | Whether the property previously received bond financing from an LHFA or bonds/Housing Credits from Florida Housing |
| `HUD Section 8 Expired Contract/Opt-Out` | Whether a HUD Section 8 contract expired or the owner opted out |
| `HUD Prepaid Mortgage` | Whether the owner prepaid a HUD multifamily insured mortgage |
| `HUD Assigned Mortgage` | Whether the mortgage was assigned to (transferred to) HUD from the original borrower |
| `RD Prepaid Mortgage` | Whether the owner prepaid a USDA Rural Development Section 515 or Section 514/516 loan |
| `Year Built/Funded` | Year built where available; otherwise, the earliest funding year. If the property did not appear in the AHI previously, obtained from FHFC records. |
| `Type of Ownership` | Owner corporation type (e.g., For-Profit, Non-Profit, Limited Dividend) |
| `Population Served` | Target population (e.g., Family, Elderly, Persons with Disabilities) |
| `Latitude` | Latitude coordinate |
| `Longitude` | Longitude coordinate |
